```python
import pandas as pd
import matplotlib.pyplot as plt

df_ex = pd.read_csv("exp/CartPole-v0_cartpole_sd1_20260223_231935/log.csv")
df_ex_rtg = pd.read_csv("exp/CartPole-v0_cartpole_rtg_sd1_20260223_233341/log.csv")
df_ex_na = pd.read_csv("exp/CartPole-v0_cartpole_na_sd1_20260223_233427/log.csv")
df_ex_rtg_na = pd.read_csv("exp/CartPole-v0_cartpole_rtg_na_sd1_20260223_233510/log.csv")

df_lb = pd.read_csv("exp/CartPole-v0_cartpole_lb_sd1_20260223_233538/log.csv")
df_lb_rtg = pd.read_csv("exp/CartPole-v0_cartpole_lb_rtg_sd1_20260223_233615/log.csv")
df_lb_na = pd.read_csv("exp/CartPole-v0_cartpole_lb_na_sd1_20260223_233640/log.csv")
df_lb_rtg_na = pd.read_csv("exp/CartPole-v0_cartpole_lb_rtg_na_sd1_20260223_233721/log.csv")
```

```python
# Plot average return vs environment steps for all four experiments
plt.figure(figsize=(8, 5))

plt.plot(df_ex['Train_EnvstepsSoFar'], df_ex['Train_AverageReturn'], label='Vanilla PG')
plt.plot(df_ex_rtg['Train_EnvstepsSoFar'], df_ex_rtg['Train_AverageReturn'], label='Reward-to-Go')
plt.plot(df_ex_na['Train_EnvstepsSoFar'], df_ex_na['Train_AverageReturn'], label='Normalize Adv')
plt.plot(df_ex_rtg_na['Train_EnvstepsSoFar'], df_ex_rtg_na['Train_AverageReturn'], label='RTG + Normalize Adv')

plt.xlabel('Number of Environment Steps')
plt.ylabel('Average Return')
plt.title('CartPole (Small Batch): Average Return vs Environment Steps')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.plot(df_lb['Train_EnvstepsSoFar'], df_lb['Train_AverageReturn'], label='LB')
plt.plot(df_lb_rtg['Train_EnvstepsSoFar'], df_lb_rtg['Train_AverageReturn'], label='LB + Reward-to-Go')
plt.plot(df_lb_na['Train_EnvstepsSoFar'], df_lb_na['Train_AverageReturn'], label='LB + Normalize Adv')
plt.plot(df_lb_rtg_na['Train_EnvstepsSoFar'], df_lb_rtg_na['Train_AverageReturn'], label='LB + RTG + Normalize Adv')

plt.xlabel('Number of Environment Steps')
plt.ylabel('Average Return')
plt.title('CartPole (Large Batch): Average Return vs Environment Steps')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
df_cheetah_baseline = pd.read_csv("exp/HalfCheetah-v4_cheetah_baseline_sd1_20260224_230105/log.csv")
df_cheetah_rLR = pd.read_csv("exp/HalfCheetah-v4_cheetah_baseline_rLR_sd1_20260225_170537/log.csv")
df_cheetah_rBG = pd.read_csv("exp/HalfCheetah-v4_cheetah_baseline_rBG_sd1_20260225_170553/log.csv")

plt.figure(figsize=(8, 5))
plt.plot(df_cheetah_baseline["Train_EnvstepsSoFar"], df_cheetah_baseline["Train_AverageReturn"], label="Baseline")
plt.plot(df_cheetah_rLR["Train_EnvstepsSoFar"], df_cheetah_rLR["Train_AverageReturn"], label="rLR")
plt.plot(df_cheetah_rBG["Train_EnvstepsSoFar"], df_cheetah_rBG["Train_AverageReturn"], label="rGS (rBG run)")
plt.xlabel("Number of Environment Steps")
plt.ylabel("Mean Reward")
plt.title("HalfCheetah: Mean Reward (Baseline vs rLR vs rGS)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
LunarLander_lambda0 = pd.read_csv("exp/LunarLander-v2_lunar_lander_lambda0_sd1_20260224_223640/log.csv")
LunarLander_lambda0d95 = pd.read_csv("exp/LunarLander-v2_lunar_lander_lambda0.95_sd1_20260224_224558/log.csv")
LunarLander_lambda0d98 = pd.read_csv("exp/LunarLander-v2_lunar_lander_lambda0.98_sd1_20260224_225001/log.csv")
LunarLander_lambda0d99 = pd.read_csv("exp/LunarLander-v2_lunar_lander_lambda0.99_sd1_20260224_225034/log.csv")
LunarLander_lambda1 = pd.read_csv("exp/LunarLander-v2_lunar_lander_lambda1_sd1_20260224_225957/log.csv")

plt.figure(figsize=(8, 5))
plt.plot(LunarLander_lambda0["Train_EnvstepsSoFar"], LunarLander_lambda0["Train_AverageReturn"], label="Lambda 0")
plt.plot(LunarLander_lambda0d95["Train_EnvstepsSoFar"], LunarLander_lambda0d95["Train_AverageReturn"], label="Lambda 0.95")
plt.plot(LunarLander_lambda0d98["Train_EnvstepsSoFar"], LunarLander_lambda0d98["Train_AverageReturn"], label="Lambda 0.98")
plt.plot(LunarLander_lambda0d99["Train_EnvstepsSoFar"], LunarLander_lambda0d99["Train_AverageReturn"], label="Lambda 0.99")
plt.plot(LunarLander_lambda1["Train_EnvstepsSoFar"], LunarLander_lambda1["Train_AverageReturn"], label="Lambda 1.0")

plt.xlabel("Number of Environment Steps")
plt.ylabel("Mean Reward")
plt.title("Lunar Lander: Mean Reward vs Environment Steps (Different Lambda Values)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
